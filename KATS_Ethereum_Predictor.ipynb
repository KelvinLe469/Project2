```python
from kats.consts import TimeSeriesData
from kats.models.prophet import ProphetModel, ProphetParams
import pandas as pd
```

```python
# Upload ETH-USD.csv to Colab
from google.colab import files

csv_file = files.upload()
```

```python
# Read the applicants_data.csv file from the Resources folder into a Pandas DataFrame
eth_data_df = pd.read_csv("ETH-USD.csv")

# Review the DataFrame
# YOUR CODE HERE

eth_data_df.tail()
```

```python
eth_data_df.columns = ['Date','Open', 'High', 'Low', 'Close', 'AdjClose', 'Volume']
eth_data_df['Date'] = pd.to_datetime(eth_data_df['Date'],infer_datetime_format=True,format='%y%m')
df_s = TimeSeriesData(time=eth_data_df['Date'], value=eth_data_df['Close'])
df_s
```

```python
# create a model param instance
params = ProphetParams(seasonality_mode='multiplicative')
# create a prophet model instance
model = ProphetModel(df_s, params)
# fit model simply by calling m.fit()
model.fit()
# make prediction for next 30 month
forecast = model.predict(steps=30, freq="MS")
forecast.head()
```
